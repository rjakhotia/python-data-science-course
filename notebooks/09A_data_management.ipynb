```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../datafiles/LR_DS1.csv")
df.head()
```

## Pivot Table

```python
df.pivot_table(index="Gender",columns="Target",values="No_OF_CR_TXNS",
                 aggfunc='mean',margins=True).round(1)
```

```python
## Pivot Table
df.pivot_table(index="Occupation",columns="Target",values="No_OF_CR_TXNS",
                 aggfunc=['count', 'mean' ],margins=True, dropna=False).round(1)
```

## Cross Tabulation

```python
pd.crosstab(index=df["Gender"], columns=df["Target"],
            values=df.No_OF_CR_TXNS,
            aggfunc=['count', 'mean'] 
            ,margins=True).round(1)
```

### Crosstab with Normalization

```python
# normalize = 'all', 'index', 'columns', True
pd.crosstab(index=df["Gender"], columns=df["Target"],
            values=df.No_OF_CR_TXNS, aggfunc='sum',
            margins=True, normalize='index').round(2)
```

## stack() function : Wide-to-Long

```python
stacked=df.stack()
stacked
```

## unstack() function: Long-to-Wide Conversion

```python
stacked.unstack()
```

## Tiling
- Breaking a continuous variable into equal-width bins — kind of like slicing a range into tiles.

```python
pd.cut(df['Age']  , bins = 4).head()
```

## One-Hot Encoding / Dummy Variables

```python
pd.get_dummies(df["Gender"], prefix = "Gender", drop_first=True).head()
```

```python
pd.get_dummies(df["Gender"], prefix = "Gender", drop_first=False).head()
```

## Factorizing values

```python
pd.factorize(df["Gender"])
```

```python
df["Gender_Factorized"] = pd.factorize(df["Gender"])[0]
df.head()
```

## Slicing / Conditional Indexing

```python
df[df["Gender_Factorized"] == 1]
```
