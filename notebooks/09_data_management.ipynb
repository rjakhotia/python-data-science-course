###  - **` Data Aggregation ( Grouping / Summarizing of data ) `**

<a href='https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/'> <img src='./images/pandas_aggregation.png' /></a>
[IMAGE REFERENCE](https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/)

````python
merge_df_123.columns
````

````python
# Step 1 - Group By: Groupby essentially splits the data 

gr_by_obj = merge_df_123.groupby(['Occupation'])
gr_by_obj.groups.keys()
````

````python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
````

````python
grouped = gr_by_obj.agg(
    {
        'Balance' : ["min", "max", "sum", "count", "mean"],
        'Age' : ["mean", "min"]
    }
)
grouped
````

````python
# grouped.columns=grouped.columns.droplevel(level = 0)
````

````python
grouped.columns = ["_".join(x) for x in grouped.columns.ravel()]

grouped.reset_index(inplace=True)
grouped
````

````python
#Renaming the column names 
grouped.columns = ["occupation", "min_bal","max_bal", 
    "sum_bal", "count_obs", "avg_bal", "avg_age", "min_age" ]

grouped
````

***
###  - **` Reorder Column Names `**

````python
grouped = grouped.reindex(columns=['occupation','count_obs', 'min_age', 'avg_age', 
                         'min_bal',  "max_bal", "sum_bal", "avg_bal" ])
grouped
````
