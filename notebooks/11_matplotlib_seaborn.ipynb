## Descriptive Statistics & Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle

LR_DF = pd.read_csv("../datafiles/LR_DS1.csv")
LR_DF.head()
```

```python
# to see plots in the notebook itself. Required for older version of Jupyter Notebook
%matplotlib inline 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Frequency Distribution

```python
## Frequency Distribution
freq = LR_DF['Occupation'].value_counts().to_frame()
freq
```

### Bar Plot

```python
freq.plot(kind='bar')
```

### Histogram

```python
#plt.hist(LR_DF['Balance'])
sns.displot(LR_DF['Balance'],kde=False, bins=10)
```

### Distribution Plot

```python
sns.displot(LR_DF['Balance'],kde=True)
```

### Box Plot

```python
bplot = sns.boxplot(y='Balance',
                 data=LR_DF)
```

```python
plt.figure(figsize=(16, 8))
bplot = sns.boxplot(y='Balance', x = 'Occupation',
                 data=LR_DF, 
                 width=0.5,
                 palette="colorblind")
```
